Os microdados da PRF
</div>

A Polícia Rodoviária Federal mantem uma base de dados abertos, no [link](https://www.gov.br/prf/pt-br/acesso-a-informacao/dados-abertos/dados-abertos), com acesso a diversas informações, basicamente agrupadas em dois tipos:

* como dados agrupados, formando o chamado Anuário, em que tabelas são disponibilizadas já com o agrupamento de informações, gerando dados já "arrumados";

* como _microdados_, que são, basicamente, os dados crus, ou seja, do jeito que são coletados, onde cada linha desse conjunto de dados se refere a um acidente ou a uma infração.

Na aula de hoje, nosso objetivo é, justamente, dar uma olhada nos microdados da PRF referente aos acidentes ocorridos em 2020 (o último ano completo) para podermos aplicar alguns dos conceitos trabalhados nessa semana, como filtragem, deleção de colunas, de linhas, e algumas outras operações possíveis de serem feitas sobre o conjunto de dados.

Para começar, vamos inicar carregando as bibliotecas que precisamos.

```python
import pandas as pd
import numpy as np
```

Naturalmente, nossa primeira tarefa será carregar os dados em um DataFrame, usando para isso o método `read_csv` da biblioteca Pandas.

Obviamente, para isso, precisamos chegar até os dados. Para o ano de 2020, os dados estão disponíveis na página da PRF, no [link](https://arquivos.prf.gov.br/arquivos/index.php/s/V1UNVfnli72JXNW). Pela disponibilidade no site, nós não conseguimos acessar os dados sem fazer download, ou seja, a opção de carregar a URL não funciona nesse caso. E para baixar basta clicar no link que aparece.

***Mas atenção: a depender das configurações de segurança de seu navegador, o download pode não ser permitido. Para resolver isso, entre nas configurações e habilite a possibilidade de fazer o download desse site!***

Ao extrair e descompactar o arquivo, nós teremos doze conjunto de dados, um referente a cada mês do ano para podermos analisar. Vamos para o do mês de fevereiro, denominado `infracoes_2020_02.csv`.

```python
Infracoes_PRF = pd.read_csv('infracoes_2020_02.csv',
                            encoding='latin1',
                            sep=';',
                            low_memory=False)
```
